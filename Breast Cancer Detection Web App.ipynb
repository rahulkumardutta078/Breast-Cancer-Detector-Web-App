# Splitting the dataset into train and test dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X=cancer_df.drop('target',axis=1)
```

```python
y=cancer_df['target']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
print("Shape of X_train is : {}".format(X_train.shape))
print("Shape of X_test is : {}".format(X_test.shape))
print("Shape of y_train is : {}".format(y_train.shape))
print("Shape of y_test is : {}".format(y_test.shape))
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_train_sc=sc.fit_transform(X_train)
X_test_sc=sc.fit_transform(X_test)
```

# Model Building

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

Finding accuracy score for each and every classification algorithms

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_classifier=LogisticRegression(random_state=51,penalty='l2')
lr_classifier.fit(X_train,y_train)

y_pred_lr=lr_classifier.predict(X_test)

accuracy_score(y_test,y_pred_lr)
```

```python
# If we pass standard scalar data to the model then accuracy score is
```

```python
lr_classifier2=LogisticRegression(random_state=51,penalty='l2')
lr_classifier2.fit(X_train_sc,y_train)

y_pred_lr_sc=lr_classifier2.predict(X_test_sc)
accuracy_score(y_test,y_pred_lr_sc)
```

## K-Nearest Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier=KNeighborsClassifier(n_neighbors=5)

knn_classifier.fit(X_train,y_train)

y_pred_knn=knn_classifier.predict(X_test)

accuracy_score(y_test,y_pred_knn)
```

```python
# If we pass standard scalar data to the model then accuracy score is
```

```python
knn_classifier2=KNeighborsClassifier(n_neighbors=5)

knn_classifier2.fit(X_train_sc,y_train)

y_pred_knn_sc=knn_classifier2.predict(X_test_sc)

accuracy_score(y_test,y_pred_knn_sc)
```

## Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB

nb_classifier=GaussianNB()
nb_classifier.fit(X_train,y_train)

y_pred_nb=nb_classifier.predict(X_test)

accuracy_score(y_test,y_pred_nb)
```

```python
# If we pass standard scalar data to the model then accuracy score is
```

```python
nb_classifier2=GaussianNB()
nb_classifier2.fit(X_train_sc,y_train)

y_pred_nb_sc=nb_classifier2.predict(X_test_sc)

accuracy_score(y_test,y_pred_nb_sc)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt_classifier=DecisionTreeClassifier(criterion='entropy',random_state=51)
dt_classifier.fit(X_train,y_train)

y_pred_dt=dt_classifier.predict(X_test)
accuracy_score(y_test,y_pred_dt)
```

```python
# If we pass standard scalar data to the model then accuracy score is
```

```python
dt_classifier2=DecisionTreeClassifier(criterion='entropy',random_state=51)
dt_classifier2.fit(X_train_sc,y_train)

y_pred_dt_sc=dt_classifier2.predict(X_test_sc)

accuracy_score(y_test,y_pred_dt_sc)
```

```python
# Now we are using gini entropy to find the accuracy for it.
```

```python
dt_classifier3=DecisionTreeClassifier(criterion='gini',random_state=51)
dt_classifier3.fit(X_train,y_train)

y_pred_dt_gini=dt_classifier3.predict(X_test)
accuracy_score(y_test,y_pred_dt_gini)
```

```python
dt_classifier4=DecisionTreeClassifier(criterion='gini',random_state=51)
dt_classifier4.fit(X_train_sc,y_train)

y_pred_dt_gini_sc=dt_classifier3.predict(X_test_sc)
accuracy_score(y_test,y_pred_dt_gini_sc)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier=RandomForestClassifier(n_estimators=20,criterion='entropy',random_state=51)

rf_classifier.fit(X_train,y_train)
y_pred_rf=rf_classifier.predict(X_test)
accuracy_score(y_test,y_pred_rf)
```

```python
# If we pass standard scalar data to the model then accuracy score is
```

```python
rf_classifier2=RandomForestClassifier(n_estimators=20,criterion='entropy',random_state=51)
rf_classifier2.fit(X_train_sc,y_train)

y_pred_rf_sc=rf_classifier2.predict(X_test_sc)

accuracy_score(y_test,y_pred_rf_sc)
```

```python
# Now we are using gini entropy to find the accuracy for it.
```

```python
rf_classifier3=RandomForestClassifier(n_estimators=20,criterion='gini',random_state=51)

rf_classifier3.fit(X_train,y_train)
y_pred_rf_gini=rf_classifier.predict(X_test_sc)
accuracy_score(y_test,y_pred_rf_gini)
```

```python
print("Accuracy Score of different Classification Algorithms...\n")

print("For Logistic Regression: {}\n".format(accuracy_score(y_test,y_pred_lr)))

print("For Logistic Regression(using Standard Scalar): {}\n".format(accuracy_score(y_test,y_pred_lr_sc)))

print("For K-Nearest Neighbors: {}\n".format(accuracy_score(y_test,y_pred_knn)))

print("For K-Nearest Neighbors(using Standard Scalar): {}\n".format(accuracy_score(y_test,y_pred_knn_sc)))

print("For Naive Bayes: {}\n".format(accuracy_score(y_test,y_pred_nb)))

print("For Naive Bayes(using Standard Scalar): {}\n".format(accuracy_score(y_test,y_pred_nb_sc)))

print("For Decision Tree(using entropy): {}\n".format(accuracy_score(y_test,y_pred_dt)))

print("For Decision Tree(using entropy and Standard Scalar): {}\n".format(accuracy_score(y_test,y_pred_dt_sc)))

print("For Decision Tree(using gini): {}\n".format(accuracy_score(y_test,y_pred_dt_gini)))

print("For Decision Tree(using gini and Standard Scalar): {}\n".format(accuracy_score(y_test,y_pred_dt_gini_sc)))


print("For Random Forest(using entropy): {}\n".format(accuracy_score(y_test,y_pred_rf)))

print("For Random Forest(using entropy and Standard Scalar): {}\n".format(accuracy_score(y_test,y_pred_rf_sc)))

print("For Random Forest(using gini): {}\n".format(accuracy_score(y_test,y_pred_rf_gini)))

```

### So, Through This accuracies we can say that Logistic Regression (using Standard Scalar) and Random Forest (using entropy) having 97 % and also having highest among other.
